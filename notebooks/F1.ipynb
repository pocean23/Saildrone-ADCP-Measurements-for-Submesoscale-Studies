```python
lon_min = single_box.longitude.min().values.item()
lon_max = single_box.longitude.max().values.item()
lat_min = single_box.latitude.min().values.item()
lat_max = single_box.latitude.max().values.item()

# Define the corners of the bounding box
polygon_lons = np.array([lon_min, lon_max, lon_max, lon_min, lon_min])
polygon_lats = np.array([lat_min, lat_min, lat_max, lat_max, lat_min])

lon_lats = polygon_lons+ 1j*polygon_lats
theta = 0.5498175376675223+0j

rot_lon_lat = lon_lats*np.exp(1j*theta)
polygon_lons_rot = rot_lon_lat.real
polygon_lats_rot = rot_lon_lat.imag
```
