```python

def regular_sep(ds):
    ds = ds.sel(time=ds.longitude.dropna(dim='time').time)
    dis = np.hstack([0,np.cumsum(gsw.distance(ds.longitude.values,ds.latitude.values))])
    ds = ds.assign_coords({'time':dis}).rename({'time':'distance'})
    regular_arr = np.arange(0,dis[-1]+1,1)
    ds = ds.drop_duplicates(dim='distance')
    ds = ds.interp(distance=regular_arr)
    return ds

def compute_sf_for_leg(ds, arr_len):
    """
    Compute structure functions for a single leg, adjusted for all cell depths.
    """
    N = ds.distance.size
    sf2_u, sf3_u,sf2_v, sf3_v, sf2_KE = [], [],[], [],[]
    delr_init = False

    for i in range(len(ds.cell_depth)):
        print(f"depth: {i}")
        u = ds.u.isel(cell_depth=i)
        v = ds.v.isel(cell_depth=i)
        sf2_depth_u, sf3_depth_u, sf2_depth_v, sf3_depth_v, sf2_depth_KE = np.zeros(N-1), np.zeros(N-1), np.zeros(N-1), np.zeros(N-1), np.zeros(N-1)
        
        for k in range(1, N):
            u_mean = u.coarsen(distance=k, boundary='trim').mean()
            v_mean = v.coarsen(distance=k, boundary='trim').mean()
            diff_u = u_mean.diff(dim='distance')
            diff_v = v_mean.diff(dim='distance')
            sf2_depth_u[k-1] = np.nanmean(diff_u**2)
            sf3_depth_u[k-1] = np.nanmean(diff_u**3)
            sf2_depth_v[k-1] = np.nanmean(diff_v**2)
            sf3_depth_v[k-1] = np.nanmean(diff_v**3)
            sf2_depth_KE[k-1] = np.nanmean(diff_u**2 + diff_v**2)
        # Extend with NaNs to match arr_len
        sf2_u.append(np.pad(sf2_depth_u, (0, arr_len - len(sf2_depth_u)), 'constant', constant_values=np.nan))
        sf3_u.append(np.pad(sf3_depth_u, (0, arr_len - len(sf3_depth_u)), 'constant', constant_values=np.nan))
        sf2_v.append(np.pad(sf2_depth_v, (0, arr_len - len(sf2_depth_v)), 'constant', constant_values=np.nan))
        sf3_v.append(np.pad(sf3_depth_v, (0, arr_len - len(sf3_depth_v)), 'constant', constant_values=np.nan))
        sf2_KE.append(np.pad(sf2_depth_KE, (0, arr_len - len(sf2_depth_KE)), 'constant', constant_values=np.nan))

    return np.array(sf2_u), np.array(sf3_u),np.array(sf2_v), np.array(sf3_v),np.array(sf2_KE)

def sf2_with_avgdata(data):
    arr_lens = [regular_sep(data.isel(legs=i)).distance.size for i in range(len(data.legs))]
    arr_len = max(arr_lens) - 1  # Adjust for diff operation reducing size by 1

    sf2_legs_u, sf3_legs_u, sf2_legs_v, sf3_legs_v, sf2_legs_KE = [], [],[], [],[]

    for l in range(len(data.legs)):
        print(f"legs: {l}")
        ds = regular_sep(data.isel(legs=l))
        sf2_u, sf3_u,sf2_v, sf3_v, sf2_KE = compute_sf_for_leg(ds, arr_len)
        sf2_legs_u.append(sf2_u)
        sf3_legs_u.append(sf3_u)
        sf2_legs_v.append(sf2_v)
        sf3_legs_v.append(sf3_v)
        sf2_legs_KE.append(sf2_KE)

    sf2_legs_u = np.stack(sf2_legs_u)
    sf3_legs_u = np.stack(sf3_legs_u)
    sf2_legs_v = np.stack(sf2_legs_v)
    sf3_legs_v = np.stack(sf3_legs_v)
    sf2_legs_KE = np.stack(sf2_legs_KE)

    delr = np.arange(1,arr_len,1)

    return sf2_legs_u, sf3_legs_u, sf2_legs_v, sf3_legs_v, sf2_legs_KE, delr
```

```python
def make_legs_pilot(ds):
    traj =np.argmax(ds.longitude.count(dim='time').values)
    lat_diff  = ds.isel(trajectory=traj).longitude.diff(dim='time').dropna(dim='time')
    idx = np.where(np.sign(lat_diff[:-1].values) != np.sign(lat_diff[1:].values))[0] #check where the SD turns from going Northward to southward or otherwise.
    red = []
    for n,i in enumerate(idx[:-3]):
        if i+6>=idx[n+1]:
            red = np.append(red,n)

    idx_leg = np.delete(idx,np.array(red,dtype=int))
    for i in range(len(idx_leg)):
        if i ==0:
            vel_ds = ds.isel(time=slice(0,idx_leg[i]))

        else:
            # print(idx_leg[i-1]+1)
            # print(idx_leg[i])
            vds1 = ds.isel(time=slice(idx_leg[i-1]+1,idx_leg[i]))
            if vds1.isel(trajectory=traj).longitude.count(dim='time')>2:
                dis_arr = gsw.distance(vds1.longitude.isel(trajectory=traj).values,vds1.latitude.isel(trajectory=traj).values)
                dist_leg = np.hstack([0,np.cumsum(dis_arr[pd.notnull(dis_arr)])])/1000
                if np.max(dist_leg>25):    # if distance more than 50km use dist//25 and apply numpy split(arr,[25,50])
                    
                    vel_ds = xr.concat([vel_ds, ds.isel(time=slice(idx_leg[i-1]+1,idx_leg[i-1]+26)),ds.isel(time=slice(idx_leg[i-1]+26,idx_leg[i]))], dim='legs')
                else:
                    vel_ds = xr.concat([vel_ds, ds.isel(time=slice(idx_leg[i-1]+1,idx_leg[i]))], dim='legs')
    #vel_ds = vel_ds.assign_coords({'legs':np.arange(len(vel_ds.legs))})
    vel_ds = vel_ds.where(vel_ds.longitude.count(dim='time')>2,drop=True)
    return vel_ds
    
```
